```python
from pandas import DataFrame
import pandas as pd
import os
import matplotlib.pyplot as plt

def get_top_songs() -> DataFrame:
    file_path = os.path.join("generated", f"2024_song_listen_time.csv")
    if os.path.isfile(file_path):
        df = pd.read_csv(file_path, encoding='latin-1')
        return df.nlargest(5, 'mins_played')
```

```python
top_songs = get_top_songs()
plt.figure(figsize=(10, 6))  # Set the figure size
bars = plt.bar(top_songs['song'], top_songs['mins_played'], color='skyblue')

# Add labels and title
plt.xlabel('Song Name')
plt.ylabel('Minutes Played')
plt.title('Top 5 Songs by Minutes Played')
plt.xticks(rotation=45)  # Rotate the song names for better readability
for bar in bars:
        yval = bar.get_height()  # Add a red marker
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 1, f"{yval:.1f}", ha='center', va='bottom', fontsize=10)
# Display the bar graph
plt.show()
```
